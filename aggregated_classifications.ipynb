```python
from pandas import DataFrame

import swasputils
```

```python
ac = swasputils.AggregatedClassifications()
```

```python
DataFrame({
    'Classification count': ac.df['Classification'].value_counts(), 
    'Label': ac.get_classification_labels(ac.df['Classification'].value_counts().index.to_series()),
}).plot(
    x='Label',
    y='Classification count',
    kind='bar',
)
```
